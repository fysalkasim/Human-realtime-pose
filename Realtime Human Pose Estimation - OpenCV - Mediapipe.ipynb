## Import Modules

```python
import cv2
import mediapipe as mp
```

```python
## initialize pose estimator
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)
```

<img src='pose_landmarks.png'>

## Pose Estimation for Video

```python
cap = cv2.VideoCapture('test_video.mp4')
while cap.isOpened():
    # read frame
    _, frame = cap.read()
    try:
        # resize the frame for portrait video
        frame = cv2.resize(frame, (350, 600))
        # convert to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # process the frame for pose detection
        
        # process the frame for pose detection
        pose_results = pose.process(frame_rgb)

# get landmark for a specific point (index 32)
        landmark32 = pose_results.pose_landmarks.landmark[32]
        x = landmark32.x
        y = landmark32.y
        z = landmark32.z

# Print the coordinates
        print("Landmark 32 - X:", x)
        print("Landmark 32 - Y:", y)
        print("Landmark 32 - Z:", z)

        pose_results = pose.process(frame_rgb)
        # print(pose_results.pose_landmarks)
        
        # draw skeleton on the frame
        mp_drawing.draw_landmarks(frame, pose_results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        # display the frame
        cv2.imshow('Output', frame)
    except:
        break
        
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
